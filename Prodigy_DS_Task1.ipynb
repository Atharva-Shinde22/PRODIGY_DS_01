# EDA and Data Visualization

```python
# Set the style for seaborn
sns.set(style="whitegrid")

# Distribution of Gender using a bar chart
plt.figure(figsize=(6,6))
sns.countplot(x='Gender', data=df, palette='viridis')
plt.title('Distribution of Gender in Employee Attrition', fontsize=16)
plt.xlabel('Gender', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.show()
```

```python
# Distribution of overtime feature using bar chart

plt.figure(figsize=(6, 6))
sns.countplot(x='OverTime', data=df, palette='Set2')
plt.title('Distribution of overtime in Employee Attrition', fontsize=16)
plt.xlabel('OverTime', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.show()
```

```python
# distribution of Marital status using Bar chart

plt.figure(figsize=(6,6))
sns.histplot(x='MaritalStatus', data=df, palette='Set2')
plt.title('Distribution of Marital Status in Employee Attrition', fontsize=16)
plt.xlabel('Marital Status', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.show()
```

```python
# Distribution of Age using a histogram

plt.figure(figsize=(8,6))
sns.histplot(df['Age'], bins=20, kde=True, color='skyblue', edgecolor='black')
plt.title('Distribution of Age in Employee Attrition', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.show()
```

```python
# Distribution of Department using a bar chart

plt.figure(figsize=(8, 6))
sns.countplot(x='Department', data=df, palette='Set2')
plt.title('Distribution of Department in Employee Attrition', fontsize=16)
plt.xlabel('Department', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.show()
```

```python
# Distribution of monthly income using histogram

plt.figure(figsize=(8, 6))
sns.histplot(df['MonthlyIncome'], bins=20, kde=True, edgecolor='black')
plt.title('Distribution of MonthlyIncome in Employee Attrition', fontsize=16)
plt.xlabel('Monthly Income', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.show()
```

```python
# Distribution of Job Role using bar chart

plt.figure(figsize=(6, 6))
sns.countplot(x='JobRole', data=df, palette='Set2')
plt.title('Distribution of Job Role in Employee Attrition', fontsize=16)
plt.xlabel('Job Role', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.xticks(rotation='vertical')
plt.show()
```

```python
# Distribution of Working Years using bar chart

plt.figure(figsize=(12, 12))
sns.countplot(x='TotalWorkingYears', data=df, palette='Set2')
plt.title('Distribution of Job Role in Employee Attrition', fontsize=16)
plt.xlabel('Total Working years', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.xticks(rotation='horizontal')
plt.show()
```

The EDA and Data Visualization provides an overview of the key variables in the HR-Employee Dataset which includes the distribution of Gender, Age, Department, Working experience, Educational background. These visualizations serves as a medium for exploring the dataset and determining the relations between the features eventually gaining insights from the data.
